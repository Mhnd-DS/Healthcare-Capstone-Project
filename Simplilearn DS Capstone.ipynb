## Handling Outliers

```python
df3_HandlingOutliers = pd.DataFrame(df2)
```

```python
df3_HandlingOutliers.shape
```

```python
for x in df3_HandlingOutliers.columns:
        plt.figure(figsize=(15,5))
        plt.subplot(121)
        plt.plot(df3_HandlingOutliers[x].values)
        plt.title(x+" BEFORE removing Outliers \n",fontsize=15)

        print('column name:', x)
        upper_limit = df3_HandlingOutliers[x].mean() + 3 * df3_HandlingOutliers[x].std()
        lower_limit = df3_HandlingOutliers[x].mean() -3 * df3_HandlingOutliers[x].std()

        outliers = df3_HandlingOutliers[(df3_HandlingOutliers[x]>upper_limit) | (df3_HandlingOutliers[x]<lower_limit)]
        print(len(outliers))

        plt.subplot(122)
        plt.plot(df2[x].values)
        plt.title(x+" AFTER removing Outliers \n",fontsize=15)
        
        df3_HandlingOutliers.drop(outliers.index , axis = 0 , inplace= True)
        print(df3_HandlingOutliers.shape[0])
```

## Scale the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
x = df3_HandlingOutliers.drop(['Outcome'] , axis = 1)
y = df3_HandlingOutliers['Outcome']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_scaled , y, stratify = y ,test_size=0.2, random_state=40) 
```

#### By the previous analysis we can use several models such as (Logistic which act as a binary classification, L1 - L2, Decision Tree, Random Forest)

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeClassifier

# We create a dict contains the models and possible desired parameters
model_params = {
    'svm': {
        'model': svm.SVC(gamma='auto'),
        'params' : {
            'C': [1,10,20,30],
            'kernel': ['rbf','linear']
        }  
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params' : {
            'n_estimators': [1,5,10,40,50,80]
        }
    },
    'logistic_regression' : {
        'model': LogisticRegression(solver='liblinear',multi_class='auto'),
        'params': {
            'C': [1,5,10,15,50]
        }
    },
    'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
    },
    'decision_tree': {
            'model': DecisionTreeClassifier(),
            'params': {
                'criterion' : ['gini','entropy'],
                'splitter': ['best','random']
            }
    }
}
```

```python
from sklearn.model_selection import GridSearchCV
scores = []

for model_name, mp in model_params.items():
    clf =  GridSearchCV(mp['model'], mp['params'], cv=20, return_train_score=False) #refit = True , verbose=2
    #clf.fit(iris.data, iris.target)
    clf.fit(x_train ,y_train)
    
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_,
    })
    
GridSearchCV_Results = pd.DataFrame(scores,columns=['model','best_score','best_params'])
GridSearchCV_Results

scores.clear()
```

```python
GridSearchCV_Results
```

## KNN Classifier as requested

```python
from sklearn.neighbors import KNeighborsClassifier

model_KNN = KNeighborsClassifier( n_neighbors= 6)

model_KNN.fit(x_train , y_train)
model_KNN.score(x_test , y_test)
```

## confusion Matrix for observation


```python
from sklearn.metrics import confusion_matrix

y_pred_KNN = model_KNN.predict(x_test)
cm = confusion_matrix(y_test , y_pred_KNN)


import seaborn as sn
sn.heatmap(cm , annot = True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

# classification report to see (f1, recall and precision)

```python
from sklearn.metrics import classification_report

print(classification_report(y_test , y_pred_KNN))
```

```python
from sklearn.svm import SVC
model_svm = SVC(C = 1 , kernel = 'linear') 
model_svm.fit(x_train , y_train)
model_svm.score(x_test , y_test)
```

```python
y_pred_SVM = model_svm.predict(x_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test , y_pred_SVM))
```

```python
from sklearn.metrics import roc_curve, auc

SVM_fpr, SVM_tpr, threshold = roc_curve(y_test, y_pred_SVM)
auc_SVM = auc(SVM_fpr, SVM_tpr)

KNN_fpr, KNN_tpr, threshold = roc_curve(y_test, y_pred_KNN)
auc_KNN = auc(KNN_fpr, KNN_tpr)

plt.figure(figsize=(5, 5), dpi=100)
plt.plot(KNN_fpr, KNN_tpr, linestyle='-', label='KNN (auc = %0.3f)' % auc_KNN)
plt.plot(SVM_fpr, SVM_tpr, marker='.', label='SVM (auc = %0.3f)' % auc_SVM)

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')

plt.legend()

plt.show()
```

### roc & AUC are important in binary classification,  for example logistic reg will make any values >.5 = 1 you need to decide the threshold

### ROC can be considered as the receiver operating characteristic
### AUC can be considered as the area under the ROC curve
